```python
# import libraries 
# from bs4 import BeautifulSoup
# import requests
import datetime
import time
import smtplib
from selenium import webdriver
from selenium.webdriver.common.by import By
import random
```

```python
#array
titles = []
addresss_1 = []
addresss_2 = []
salarys = [] 
posted_dates = [] 
companys = []
def remove_between_brackets(text):
    result = []
    inside_brackets = False
    for char in text:
        if char == '<':
            inside_brackets = True
            result.append(",")
        elif char == '>':
            inside_brackets = False
        elif inside_brackets == False:
            result.append(char)
    result = ''.join(result[1:])
    result_1 = result.split(",")
    return result_1[:2]



def main(num):
    # Khởi tạo WebDriver (sử dụng ChromeDriver trong ví dụ này)
    # Generate the URL
    driver = webdriver.Chrome()
    URLs = 'https://www.topcv.vn/tim-viec-lam-it-phan-mem-c10026?sort=new&page=' + str(num) + '&sba=1'
    print(URLs)
    driver.get(URLs)
    # Retrieve elements
    spans = driver.find_elements(By.CSS_SELECTOR, "span[data-toggle='tooltip']")
    soup_adds = driver.find_elements(By.CSS_SELECTOR, "label.address")
    soup_dates = driver.find_elements(By.CSS_SELECTOR, "label.address.mobile-hidden")
    soup_salarys = driver.find_elements(By.CSS_SELECTOR, "label.title-salary")
    soup_companys = driver.find_elements(By.CSS_SELECTOR, "a[data-toggle='tooltip'][target='_blank']")
    #company
    for soup_company in soup_companys[:-3]:
        soup_company = soup_company.text
        if soup_company != "":
            companys.append(soup_company)
    #salary
    for soup_salary in soup_salarys:
        soup_salary = soup_salary.text
        salarys.append(soup_salary)
    #date
    for soup_date in soup_dates:
        soup_date = soup_date.text
        posted_dates.append(soup_date)
    #title
    for span in spans:
        title = span.text
        if '<' not in title:
            titles.append(title)
    #address
    for soup_add in soup_adds:
        try:
            address = soup_add.get_attribute("data-original-title")
            if address:
                cleaned_text = remove_between_brackets(address)
                addresss_1.append(cleaned_text[0])
                addresss_2.append(cleaned_text[1])
        except KeyError:
            continue
    driver.quit()
if __name__ == "__main__":
    num = 1
    while num <= 55: # Fixed number on 8/25/2024
        main(num)
        num += 1
        delay = random.uniform(4, 8) # Random time sleep
        time.sleep(delay) 
```

```python
import pandas as pd
import re
from openpyxl import load_workbook
from datetime import date



d = {"Title":[], "Skill 1":[], "Skill 2":[], "Skill 3":[], "Skill 4":[], "Company":[], "Address 1":[], "Address 2":[], "Post date":[], "Salary": []}

def field(string):
    # ITViec
    desired_technologies_1 = [
        "ABAP", "Agile", "Android", "Angular", "AngularJS", "ASP.NET", "Automation Test", "AWS",
        "Azure", "Blockchain", "Bridge Engineer", "Business Analyst", "Business Intelligence", "C#",
        "C++", "C language", "Cloud", "Cocos", "Computer Vision", "CSS", "Dart", "Data Analyst",
        "Database", "Designer", "DevOps", "DevSecOps", "Django", "Embedded", "Embedded C", "English",
        "ERP", "Flutter", "Games", "Golang", "HTML5", "iOS", "IT Support", "J2EE", "Japanese", "Java",
        "JavaScript", "JQuery", "JSON", "Kotlin", "Laravel", "Linux", "Magento", "Manager", "MongoDB",
        "MVC", "MySQL", ".NET", "Networking", "NodeJS", "NoSQL", "Objective C", "OOP", "Oracle", "PHP",
        "PostgreSql", "PQA", "Presale", "Product Designer", "Product Manager", "Product Owner",
        "Project Manager", "Python", "QA QC", "ReactJS", "React Native", "Ruby", "Ruby on Rails",
        "Salesforce", "Scala", "Scrum", "Security", "Software Architect", "Solidity", "Solution Architect",
        "Spring", "SQL", "Swift", "System Admin", "System Engineer", "Team Leader", "Tester", "TypeScript",
        "UI-UX", "Unity", "VueJS", "Wordpress"
    ]
    # Duy
    desired_technologies_2 = [
        "Mobile", "Game", "AI", "Web", "Leader", "Data"
    ]
    desired_technologies = desired_technologies_1 + desired_technologies_2
    string = string.lower()
    extracted_technologies = [tech for tech in desired_technologies if tech.lower() in string]
    return extracted_technologies[:4]
    
for title, company, address_1, address_2, post_date, salary in zip(titles, companys, addresss_1, addresss_2, posted_dates, salarys):
    abc = field(title)
    chieu_dai = len(abc)
    flag = False
    for x in range(4):
        if x < chieu_dai:
            d['Skill '+ str(x + 1)].append(abc[x])
        else:
            d['Skill '+ str(x + 1)].append('')
    d['Title'].append(title)
    d['Company'].append(company)
    d['Address 1'].append(address_1)
    d['Address 2'].append(address_2)
    d['Post date'].append(post_date)
    d['Salary'].append(salary)
    time.sleep(0.004)
    
# Name file csv
today = date.today()
file_path = "TopCV" + " " + str(today) + ".csv"
TopCV_df = pd.DataFrame.from_dict(d)
TopCV_df.to_csv(file_path, header=True, index=False, encoding='utf-8')
```
